```r
#estimating variance of s1, accounting for missing data
MLE<-function(var, s1, z){
    estimate<-0
    for(i in s1){
        estimate <- estimate+log(dnorm(i, mean=0, sd=sqrt(var)))
    }
    estimate <- estimate + (M - length(s1)) * log(1-2*pnorm(z, mean = 0, sd = sqrt(var)))

    return(estimate)
}

#variance explained by genetics
expected_mean_ratio<-function(sigma_g_2, maxVar, n1, n2){
    sqrt(n1)*sqrt(n2)*sigma_g_2/maxVar
}
#estimate variance components
estimate_parameters <- function(s1, s2, n1, n2, z){
    #sigma_g, sigma_c1, sigma_c2 initial values
    parameters <- c(1e-28, 1e-28, 1e-28)

    ######### estimate total variance in s1  #############
    max<- -Inf
    maxVar<- -Inf
    for(i in seq(from=0.1, to=100, by=1)){
        temp<-MLE(i, s1, z)
        print(temp)
        if (temp>max) {
            max<-temp
            maxVar<-i
        }
    }
    print(maxVar)

    ######### estimate sigma_g  #############
    min_rms <- Inf
    var_g_est2 <- 1e-28


    for(i in seq(from=1e-28,to=.1, by=0.001)){
        ratio<-expected_mean_ratio(i, maxVar, n1, n2)
        expected_s2 <- s1*ratio
        cur_rms <- sqrt(sum((expected_s2-s2)^2))
        if (cur_rms < min_rms){
                min_rms<- cur_rms
                var_g_est2 <- i
        }
    }
    #set sigma_g parameter
    parameters[1] <- var_g_est2
    print(var_g_est2)
    
    ######### estimate sigma_c1 #############
    c1_est2<-(maxVar-1-n1*var_g_est2)/n1
    

    if(c1_est2<0){
        c1_est2<-1e-28
        parameters[1] <- maxVar-1 #c1_est2 sampling error, so set to zero
        print("sampling error")
    }
    #set sigma_c1 parameter
    parameters[2] <- c1_est2
    print(c1_est2)
    
    ######### estimate sigma_c2 #############
    max<- - Inf
    c2_est<- 1e-28
    stats <- cbind(s1, s2)
    lik <- c()
    for(i in seq(from=1e-28,to=.1, by=0.001)){
        temp <- MLE_joint_probability(var_g_est2, c1_est2, i, n1, n2, stats, z)
        if(temp>max){
            max<-temp
            c2_est<-i
        }
        lik <- c(lik, temp)
    }

    #set sigma_c2 parameter
    parameters[3] <- c2_est
    print(c2_est)

    return(c(var_g_est2,c1_est2, c2_est))
}


#doesn't work well b/c likelihood just increases as you increase the variance
estimate_parameters_joint <- function(s1, s2, n1, n2){
    stats <- cbind(s1, s2)
    print(dim(stats))
    #sigma_g, sigma_c1, sigma_c2 initial values
    parameters <- c(1e-28, 1e-28, 1e-28)
    loglik <- -Inf
    for(var_g in seq(from=1e-28,to=.005, by=0.001)){
        for(var_c1 in seq(from=1e-28,to=.005, by=0.001)){
        
            for(var_c2 in seq(from=1e-28,to=.005, by=0.001)){
                ll <- MLE_joint_probability(var_g, var_c1, var_c2, n1, n2, stats)
                if(ll > loglik){
                    parameters[1] <- var_g
                    parameters[2] <- var_c1
                    parameters[3] <- var_c2
                }
            }
        }
    }
  

    return(parameters)
}
MLE_joint_probability<-function(var_g, var_c1, var_c2, n1, n2, stats, z){
    cov_matrix=matrix(data=NA, nrow=2, ncol=2)
    cov_matrix[1,1]=n1*var_g+n1*var_c1+1
    cov_matrix[1,2]=sqrt(n1)*var_g*sqrt(n2)
    cov_matrix[2,1]=sqrt(n1)*var_g*sqrt(n2)
    cov_matrix[2,2]=n2*var_g+n2*var_c2+1

    mean_matrix=matrix(data=NA, nrow=2, ncol=1)
    mean_matrix[1,1]=0
    mean_matrix[2,1]=0

    estimate<-0

    mean_matrix <- data.matrix(mean_matrix)
    
    prob<-dmvnorm(x=stats, mean = c(0,0), sigma = cov_matrix, log = TRUE)

    for(i in prob){
        estimate<-estimate + i
    }
    estimate = estimate + 0 * log(1-2*pmvnorm(lower=-Inf,upper=z, mean=c(0,0),sigma = cov_matrix))
    return(estimate)
}



predict_replication <- function(mean, sd, z){
    #calculate predicted replication rate
    lower <-  pnorm(z, mean, sd)
    upper <- 1- pnorm(-z, mean, sd)
    return(lower + upper)
}


#calculate conditional probability of s2 > t | s1 = x under model with no confounding
calc_conditional_no_confounding <- function(s1, n1, n2, var_g, z){


    mean <- (sqrt(n1)*sqrt(n2)*var_g)/(1+n1*var_g) *s1
    sd <- sqrt(n2*var_g+1-((n1*var_g*var_g*n2)/(n1*var_g+1)))
    lower <-  pnorm(z, mean, sd)
    upper <- 1- pnorm(-z, mean, sd)
    return(lower + upper)
}
#calculate conditional probability of s2 > t | s1 = x under model with confounding
calc_conditional_with_confounding <- function(s1, n1, n2, var_g, var_c1, var_c2, z){

    mean <- (sqrt(n1)*sqrt(n2)*var_g)/(1+n1*var_g+n1*var_c1) *s1
    sd <- sqrt(n2*var_g+n2*var_c2+1-((n1*var_g^2*n2)/(n1*var_g+n1*var_c1+1)))
    lower <-  pnorm(z, mean, sd)
    upper <- 1-pnorm(-z, mean, sd)
    return(lower + upper)

}

#calculate predicted replication rate under model with no confounding
predict_no_confounding <- function(data, z, sigma_g){

    predicted_replication <- 0
    for(i in 1:nrow(data)){
        predicted_replication <- predicted_replication +  calc_conditional_no_confounding(data[i,1], n1, n2, sigma_g, z)

    }

    return(predicted_replication/nrow(data))
}

#calculate predicted replication rate under model with confounding
predict_with_confounding <- function(data, z, sigma_g, sigma_c1, sigma_c2){

    predicted_replication <- 0
    for(i in 1:nrow(data)){
         predicted_replication <- predicted_replication +  calc_conditional_with_confounding(data[i,1], n1, n2, sigma_g, sigma_c1, sigma_c2, z)

    }

    return(c(predicted_replication/nrow(data), sigma_g, sigma_c1, sigma_c2))
}

calcReplication <- function(data, z){
    count <- 0
    for(i in 1:nrow(data)){
        if(abs(data[i,2])>abs(z)){
            count  <- count + 1
        }
    }
    return(count/nrow(data))
}
run_analysis <- function(f, n1, n2){

    d <- read.table(f, skip=1, header=TRUE)
    threshold1 <- 0.05/nrow(d)
    z1 <- qnorm(threshold1/2)
    sig <- subset(d, abs(d$s1)> abs(z1))
    
#   stopifnot(nrow(sig)>1) #cannot apply method if there is not at least one significant variant

    #z-score significance threshold for replication study
    threshold2 <- 0.05/nrow(sig)
    z2 <- qnorm(threshold2/2)
    #true replication rate
    r_true <- calcReplication(sig, z2)


    #estimate parameters
    s1 <- sig[,1]
    s2 <- sig[,2]

    parameters <- estimate_parameters(s1, s2, n1, n2, z2)
    sigma_g <- parameters[1]
    sigma_c1 <- parameters[2]
    sigma_c2 <- parameters[3]

    #predicted replication rate with no confounding
    pr_no_confounding <- predict_no_confounding(sig, z2, sigma_g)

    #predicted replication rate with confounding, sigma_g1, sigma_c1, and sigma_c2
    pr_with_confounding <- predict_with_confounding(sig, z2, sigma_g, sigma_c1, sigma_c2)


    #output predicted replication rates and parameters
    #write.table(c(r_true, pr_no_confounding, pr_with_confounding), file=outfile, row.names=F, col.names=F, quote=F)

    return(c(sigma_g, sigma_c1, sigma_c2))
}
```

```r
#estimating variance of s1, accounting for missing data
MLE_trunc<-function(var, s1, z){
    estimate<-0
    for(i in s1){
        if(sign(i)==1){
            estimate<-estimate+log(dtruncnorm(i, mean=0, sd=sqrt(var), a=abs(z), b=Inf))
        } else{
            estimate<-estimate+log(dtruncnorm(i, mean=0, sd=sqrt(var), a=-Inf, b=z))
        }
        print(estimate)
        
    }

    return(estimate)
}

#estimating variance of s1, accounting for missing data
MLE_norm<-function(var, s1, z){
    estimate<-0
    for(i in s1){
        estimate<-estimate+log(dnorm(i, mean=0, sd=sqrt(var)))

        
    }

    return(estimate)
}

#variance explained by genetics
expected_mean_ratio<-function(sigma_g_2, maxVar, n1, n2){
    sqrt(n1)*sqrt(n2)*sigma_g_2/maxVar
}

compute_var_g<- function(maxVar, maxVar2, n1, n2){
    var_g <- 1e-28
    ll <- -Inf
    for(i in seq(from=.001, to=.2, by=.001)){
        cov_matrix=matrix(data=NA, nrow=2, ncol=2)
        cov_matrix[1,1]=maxVar
        cov_matrix[1,2]=sqrt(n1)*i*sqrt(n2)
        cov_matrix[2,1]=sqrt(n1)*i*sqrt(n2)
        cov_matrix[2,2]=maxVar2
        prob<-dmvnorm(x=stats, mean = c(0,0), sigma = cov_matrix, log = TRUE)
        prob <- sum(prob)
        if(prob > ll){
            ll <- prob
            var_g <- i
        }
    }
}

#estimate variance components
estimate_parameters_trunc <- function(s1, s2, n1, n2, z){
    #sigma_g, sigma_c1, sigma_c2 initial values
    parameters <- c(1e-28, 1e-28, 1e-28)

    ######### estimate total variance in s1  #############
    max<- -Inf
    maxVar<- -Inf
    for(i in seq(from=0.1, to=100, by=1)){
        temp<-MLE_trunc(i, s1, z)
        if (temp>max) {
            max<-temp
            maxVar<-i
        }
    }
    print(maxVar)
    
    ######### estimate total variance in s2  #############
#    max2<- -Inf
#    maxVar2<- -Inf
#    for(i in seq(from=0.1, to=10000, by=10)){
#        temp<-MLE_trunc(i, s2, z)
#        if (temp>max2) {
#            max2<-temp
#            maxVar2<-i
#        }
#    }
#    print(maxVar2)


    ######### estimate sigma_g  #############
    min_rms <- Inf
    var_g_est2 <- 1e-28
    for(i in seq(from=1e-28,to=.1, by=0.001)){
        ratio<-expected_mean_ratio(i, maxVar, n1, n2)
        expected_s2 <- s1*ratio
        cur_rms <- sqrt(sum((expected_s2-s2)^2))
        if (cur_rms < min_rms){
                min_rms<- cur_rms
                var_g_est2 <- i
        }
    }
    
    #set sigma_g parameter
#    parameters[1] <- compute_var_g(maxVar, maxVar2, n1, n2)
#    print(var_g_est2)
    
    ######### estimate sigma_c1 #############
    c1_est2<-(maxVar-1-n1*var_g_est2)/n1
    

    if(c1_est2<0){
        c1_est2<-1e-28
        parameters[1] <- maxVar-1 #c1_est2 sampling error, so set to zero
        print("sampling error")
    }
    #set sigma_c1 parameter
    parameters[2] <- c1_est2
    print(c1_est2)
    
    ######### estimate sigma_c2 #############
    max<- - Inf
    c2_est<- 1e-28
    stats <- cbind(s1, s2)
    lik <- c()
    for(i in seq(from=1e-28,to=.1, by=0.001)){
        temp <- MLE_joint_probability(var_g_est2, c1_est2, i, n1, n2, stats, z)
        if(temp>max){
            max<-temp
            c2_est<-i
        }
        lik <- c(lik, temp)
    }

    #set sigma_c2 parameter
    parameters[3] <- c2_est
    print(c2_est)

    return(c(var_g_est2,c1_est2, c2_est))
}


#doesn't work well b/c likelihood just increases as you increase the variance
estimate_parameters_joint <- function(s1, s2, n1, n2){
    stats <- cbind(s1, s2)
    print(dim(stats))
    #sigma_g, sigma_c1, sigma_c2 initial values
    parameters <- c(1e-28, 1e-28, 1e-28)
    loglik <- -Inf
    for(var_g in seq(from=1e-28,to=.005, by=0.001)){
        for(var_c1 in seq(from=1e-28,to=.005, by=0.001)){
        
            for(var_c2 in seq(from=1e-28,to=.005, by=0.001)){
                ll <- MLE_joint_probability(var_g, var_c1, var_c2, n1, n2, stats)
                if(ll > loglik){
                    parameters[1] <- var_g
                    parameters[2] <- var_c1
                    parameters[3] <- var_c2
                }
            }
        }
    }
  

    return(parameters)
}
MLE_joint_probability<-function(var_g, var_c1, var_c2, n1, n2, stats, z){
    cov_matrix=matrix(data=NA, nrow=2, ncol=2)
    cov_matrix[1,1]=n1*var_g+n1*var_c1+1
    cov_matrix[1,2]=sqrt(n1)*var_g*sqrt(n2)
    cov_matrix[2,1]=sqrt(n1)*var_g*sqrt(n2)
    cov_matrix[2,2]=n2*var_g+n2*var_c2+1

    mean_matrix=matrix(data=NA, nrow=2, ncol=1)
    mean_matrix[1,1]=0
    mean_matrix[2,1]=0

    estimate<-0

    mean_matrix <- data.matrix(mean_matrix)
    
    prob<-dmvnorm(x=stats, mean = c(0,0), sigma = cov_matrix, log = TRUE)

    for(i in prob){
        estimate<-estimate + i
    }
    estimate = estimate + 0 * log(1-2*pmvnorm(lower=-Inf,upper=z, mean=c(0,0),sigma = cov_matrix))
    return(estimate)
}



predict_replication <- function(mean, sd, z){
    #calculate predicted replication rate
    lower <-  pnorm(z, mean, sd)
    upper <- 1- pnorm(-z, mean, sd)
    return(lower + upper)
}


#calculate conditional probability of s2 > t | s1 = x under model with no confounding
calc_conditional_no_confounding <- function(s1, n1, n2, var_g, z){


    mean <- (sqrt(n1)*sqrt(n2)*var_g)/(1+n1*var_g) *s1
    sd <- sqrt(n2*var_g+1-((n1*var_g*var_g*n2)/(n1*var_g+1)))
    lower <-  pnorm(z, mean, sd)
    upper <- 1- pnorm(-z, mean, sd)
    return(lower + upper)
}
#calculate conditional probability of s2 > t | s1 = x under model with confounding
calc_conditional_with_confounding <- function(s1, n1, n2, var_g, var_c1, var_c2, z){

    mean <- (sqrt(n1)*sqrt(n2)*var_g)/(1+n1*var_g+n1*var_c1) *s1
    sd <- sqrt(n2*var_g+n2*var_c2+1-((n1*var_g^2*n2)/(n1*var_g+n1*var_c1+1)))
    lower <-  pnorm(z, mean, sd)
    upper <- 1-pnorm(-z, mean, sd)
    return(lower + upper)

}

#calculate predicted replication rate under model with no confounding
predict_no_confounding <- function(data, z, sigma_g){

    predicted_replication <- 0
    for(i in 1:nrow(data)){
        predicted_replication <- predicted_replication +  calc_conditional_no_confounding(data[i,1], n1, n2, sigma_g, z)

    }

    return(predicted_replication/nrow(data))
}

#calculate predicted replication rate under model with confounding
predict_with_confounding <- function(data, z, sigma_g, sigma_c1, sigma_c2){

    predicted_replication <- 0
    for(i in 1:nrow(data)){
         predicted_replication <- predicted_replication +  calc_conditional_with_confounding(data[i,1], n1, n2, sigma_g, sigma_c1, sigma_c2, z)

    }

    return(c(predicted_replication/nrow(data), sigma_g, sigma_c1, sigma_c2))
}

calcReplication <- function(data, z){
    count <- 0
    for(i in 1:nrow(data)){
        if(abs(data[i,2])>abs(z)){
            count  <- count + 1
        }
    }
    return(count/nrow(data))
}
run_analysis_trunc <- function(f, n1, n2){

    d <- read.table(f, skip=1, header=TRUE)
    threshold1 <- 0.05/nrow(d)
    z1 <- qnorm(threshold1/2)
    sig <- subset(d, abs(d$s1)> abs(z1))
    
#   stopifnot(nrow(sig)>1) #cannot apply method if there is not at least one significant variant

    #z-score significance threshold for replication study
    threshold2 <- 0.05/nrow(sig)
    z2 <- qnorm(threshold2/2)
    #true replication rate
    r_true <- calcReplication(sig, z2)


    #estimate parameters
    s1 <- sig[,1]
    s2 <- sig[,2]

    parameters <- estimate_parameters_trunc(s1, s2, n1, n2, z2)
    sigma_g <- parameters[1]
    sigma_c1 <- parameters[2]
    sigma_c2 <- parameters[3]

    #predicted replication rate with no confounding
    pr_no_confounding <- predict_no_confounding(sig, z2, sigma_g)

    #predicted replication rate with confounding, sigma_g1, sigma_c1, and sigma_c2
    pr_with_confounding <- predict_with_confounding(sig, z2, sigma_g, sigma_c1, sigma_c2)


    #output predicted replication rates and parameters
    #write.table(c(r_true, pr_no_confounding, pr_with_confounding), file=outfile, row.names=F, col.names=F, quote=F)

    return(c(sigma_g, sigma_c1, sigma_c2))
}
```
