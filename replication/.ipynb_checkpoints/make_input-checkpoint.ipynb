```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

# Format replication modeling input

```python
ox = pd.read_table("~/project-zarlab/CFW/MegaAnalysis_CFW_2019_Summary_Statistics/Flint_July14/OXPeaks160620.txt")
ox.head()
```

```python
sig = ox[ox["ox.logp"] > 5.675524]
sig.shape
```

```python
ox["phenotype"] = [p[:-20] for p in ox["phenotype"]]
ox.head()
```

```python
ox_z = []
uc_z = []
for i in ox.index:
    p1 = 10** (- ox.loc[i, "ox.logp"])
    z1 = norm.ppf(p1)
    if ox.loc[i, "ox.beta"] > 0:
        z1 = -1 * z1
        
    p2 = 10** (- ox.loc[i, "uc.logp"])
    if np.isnan(p2):
        ox_z.append(np.nan)
        uc_z.append(np.nan)
        continue
    z2 = norm.ppf(p2)
    if ox.loc[i, "uc.beta"] > 0:
        z2 = -1 * z2
    
    ox_z.append(z1)
    uc_z.append(z2)
    
```

```python
ox["z1"] = ox_z
ox["z2"] = uc_z
ox = ox.dropna(subset=["z1", "z2"])
ox.shape
```

```python
sig = ox[ox["ox.logp"] > 5.675524]
sig.shape
```

```python
sig.head()
```

```python
keep_var = pd.read_table("/u/home/j/jzou1115/project-zarlab/CFW/confidence_interval_pipeline_out_pruned_ox_redo/qtls_independent.txt", header=None)[0]
```

```python
sig["qtlid"] = [sig.loc[i, "phenotype"]+"_"+sig.loc[i, "marker"] for i in sig.index]
sig = sig[sig["qtlid"].isin(keep_var)]
sig.shape
```

```python
datadir = "/u/home/j/jzou1115/project-zarlab/CFW/replication_pruned/input_phenotypes_pooled_2"
for p in set(sig["phenotype"]):
    print(p)
    sub = sig[sig["phenotype"]==p]
    if sub.shape[0] >= 2:
        outfile = datadir+"/"+p+"_ox_uc.txt"
        print(outfile)
        sub.to_csv(outfile, columns=["z1", "z2"], header=["s1", "s2"], sep="\t", index_label=False)
```
